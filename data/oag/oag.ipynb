```python
%load_ext lab_black

import pandas as pd


def preview(dataframe, head=5):
    display(dataframe.head(head))
    print("This dataframe has " + str(len(dataframe)) + " observations.")
```

```python
oag = pd.read_parquet("/data/data/open_dataset/oag/2018.parquet")
```

```python
preview(oag)
```

```python
flights = (
    oag.groupby(["FlightNo1", "Carrier1"])
    .count()["Km"]
    .reset_index()
    .rename(
        columns={
            "Km": "flights",
        }
    )
)
preview(flights)
```

```python
flights = (
    oag.groupby(["FlightNo1"])
    .count()["Km"]
    .reset_index()
    .rename(
        columns={
            "Km": "flights",
        }
    )
)
preview(flights)
```

```python
flights = (
    oag.groupby(["GeneralAcft", "DepAirport", "ArrAirport"])
    .count()["Km"]
    .reset_index()
    .rename(
        columns={
            "Km": "flights",
        }
    )
)
preview(flights)
```

```python
flights = (
    oag.groupby(["GeneralAcft", "GeneralAcftName", "DepAirport", "ArrAirport"])
    .count()["Km"]
    .reset_index()
    .rename(
        columns={
            "Km": "flights",
        }
    )
)
preview(flights)
```

```python
oag.query("GeneralAcft=='100'")[["Carrier1", "FlightNo1", "DepAirport", "ArrAirport"]]
```

```python
arr_airports = ["ADE", "AMS", "AZD", "BND"]
dep_airports = ["JIB", "BRU", "MUC", "BND"]
flights.query(
    f"GeneralAcft=='100' and ArrAirport in {arr_airports} and DepAirport in {dep_airports}"
)
```
